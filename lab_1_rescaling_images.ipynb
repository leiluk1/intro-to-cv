## Task 1
Download a small color digital image from the internet. 

Write a good algorithm that rescales this image to half its size, in both axis (width and height). Print the two images.


```python
# import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# you can use your own image here
img = plt.imread('image.jpg')
plt.imshow(img)
plt.title(f'An original image of shape {img.shape}',
          fontsize=10)
plt.show()
```

After the image is successfully loaded, let's proceed with an algorithm that rescales this image to half its original size.

To produce a new image scaled down by 2 compared to the original image, I will keep every second pixel of the original image in both dimensions. To do this, first we need to create a new empty image with the corresponding dimensions, then iterate over each pixel of the created image and assign it the corresponding value from the original image.

```python
def rescale_img_down(original_img):
    # get the width and height of the original image
    height, width = original_img.shape[:2]

    # calculate the width and height of a new rescaled image
    new_width, new_height = width // 2, height // 2

    # create an empty array with the dimensions corresponding to the rescaled image
    rescaled_img = np.zeros((new_height, new_width, 3)).astype('uint8')

    # iterate over each pixel of a new image and assign the corresponding pixel value from the original image
    for i in range(new_height):
        for j in range(new_width):
            rescaled_img[i, j] = img[i * 2, j * 2]
    return rescaled_img
```

```python
# rescale the original image
rescaled_img = rescale_img_down(img)

# display the results
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 12))
fig.suptitle('The results of rescaling an image to half its size', y=0.73)
ax1.imshow(img)
ax1.set_title(f'Original image of shape {img.shape}',
              fontsize=10)
ax2.imshow(rescaled_img)
ax2.set_title(f'Rescaled image of shape {rescaled_img.shape}',
              fontsize=10)
plt.show()
```

## Task 2

Write a good algorithm that rescales this image to double its size, in both axis (width and height). Print
the two images.

To rescale the original image to double its size, each pixel of the original image is replicated four times (on three more pixel neighbour positions of the rescaled image) in order to increase the image's scale by two times. To do so, we need also create a new empty image and place the corresponding pixel value from the original image.

```python
def rescale_img_up(original_img):
    # get the width and height of the original image
    height, width = original_img.shape[:2]

    # calculate the width and height of a new rescaled image
    new_width, new_height = width * 2, height * 2

    # create an empty array with the dimensions corresponding to the rescaled image
    rescaled_img = np.zeros((new_height, new_width, 3)).astype('uint8')

    # iterate over each pixel of a new image and assign the corresponding pixel value from the original image
    for i in range(new_height):
        for j in range(new_width):
            rescaled_img[i, j] = original_img[i // 2, j // 2]
    return rescaled_img
```

```python
# rescale the original image
rescaled_img = rescale_img_up(img)

# display the results
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 12))
fig.suptitle('The results of rescaling an image to double its size', y=0.73)
ax1.imshow(img)
ax1.set_title(f'Original image of shape {img.shape}',
              fontsize=10)
ax2.imshow(rescaled_img)
ax2.set_title(f'Rescaled image of shape {rescaled_img.shape}',
              fontsize=10)
plt.show()
```
